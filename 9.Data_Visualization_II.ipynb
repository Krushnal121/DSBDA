```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

tdata=pd.read_csv(r"./titanic9.csv")
```

```python
tdata.head()
tdata.info()
tdata.isnull().sum()
```

```python
tdata['Age'] = tdata['Age'].fillna(np.mean(tdata['Age']))
tdata['Cabin'] = tdata['Cabin'].fillna(tdata['Cabin'].mode()[0])
tdata['Embarked'] = tdata['Embarked'].fillna(tdata['Embarked'].mode()[0])
```

```python
sns.countplot( x ="Sex", data = tdata)
plt.show()
```

```python
sns.boxplot(x="Sex",y="Age", data = tdata, hue='Survived', color = "red")
plt.title("Distribution of age and gender wrt survival rate")
plt.show()
```
